```python
#dropping unnamed columns

data.drop(columns='Unnamed: 0',inplace=True)
data_corr.drop(columns='Unnamed: 0',inplace=True)
```

### Split Into Train And Test Data 

```python
#Lets Split our data

X = data.drop(columns='offset_fault')
y = data['offset_fault']
X_corr = data_corr.drop(columns='offset_fault')
y_corr = data_corr['offset_fault']

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0,stratify=y)
X_train_corr,X_test_corr,y_train_corr,y_test_corr = train_test_split(X_corr,y_corr,test_size=0.25,random_state=0,stratify=y_corr)
```

### Scaling 

```python
# Lets go on to feature scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled_corr = scaler.fit_transform(X_train_corr)
X_test_scaled_corr = scaler.transform(X_test_corr)
```

We used MinMaxScaler cause some of the feature distributions are not normal, StandardScaler algorithm will try to change all the distributions to normal, but we don't want that, even if that is generally more pleasurable for ml algorithms, changing a feature distribution is not very pleasurable, so we want to preserve that, in summary StandarScaler is preferred when features are on same scales, which is not our case

```python
test_set['Datetime'] = pd.to_datetime(test_set["Datetime"])

day = test_set['Datetime'].dt.day
month = test_set['Datetime'].dt.month
year = test_set['Datetime'].dt.year
```

```python
test_set['day'] = day
test_set['month'] = month
test_set['year'] = year
```

```python
test_set.columns = test_set.columns.str.lower()
test_set.rename(columns ={'sensor1_pm2.5':'sensor1','sensor2_pm2.5':'sensor2'},inplace=True) #renaming columns to make life easier
```

```python
test_set.loc[test_set.sensor1.isna()==True,['sensor1']] = test_set['sensor1'].median()
test_set.loc[test_set.sensor1.isna()==True,['sensor2']] = test_set['sensor2'].median()

test_set.loc[test_set.temperature.isna()==True,['temperature']] = test_set['temperature'].mean()
test_set.loc[test_set.relative_humidity.isna()==True,['relative_humidity']] = test_set['relative_humidity'].mean()
```

```python
test_set['dew_point'] = (243.04 * (np.log(test_set['relative_humidity']/100) + ((17.625 * test_set['temperature'])/(243.04 + test_set['temperature'])))) / (17.625 - (np.log(test_set['relative_humidity']/100) + ((17.625 * test_set['temperature'])/(243.04 + test_set['temperature']))))
test_set['partial_pressure'] = test_set['relative_humidity']/100 * 6.105 * np.exp((17.27 * test_set['temperature'])/(237.7 + test_set['temperature']))
test_set['svp'] = 6.11 * 10**(7.5*test_set['temperature']/(237.3 + test_set['temperature']))
test_set['heat_index'] = -42.379 + 2.04901523 * test_set.temperature + 10.14333127 * test_set.relative_humidity - 0.22475541 * test_set.temperature * test_set.relative_humidity - 6.83783 * 10**-3 * test_set.temperature**2 - 5.481717 * 10**-2 * test_set.relative_humidity**2 + 1.22874 * 10**-3 * test_set.temperature**2 * test_set.relative_humidity + 8.5282 * 10**-4 * test_set.temperature * test_set.relative_humidity**2 - 1.99 * 10**-6 * test_set.temperature**2 * test_set.relative_humidity**2
test_set['THI'] = 0.81 * test_set.temperature + 0.01 * test_set.relative_humidity * (0.99 * test_set.temperature - 14.3) + 46.3


test_set['sensor1_rolling30'] = test_set['sensor1'].rolling(30).mean()
test_set['sensor1_rolling60'] = test_set['sensor1'].rolling(60).mean()
test_set['sensor2_rolling30'] = test_set['sensor2'].rolling(30).mean()
test_set['sensor2_rolling60'] = test_set['sensor2'].rolling(60).mean()
```

```python
test_set['sensor1_rolling30'].where(test_set['sensor1_rolling30'].isna()==False,test_set['sensor1_rolling30'].min(),inplace=True)
test_set['sensor1_rolling60'].where(test_set['sensor1_rolling60'].isna()==False,test_set['sensor1_rolling60'].min(),inplace=True)
test_set['sensor2_rolling30'].where(test_set['sensor2_rolling30'].isna()==False,test_set['sensor2_rolling30'].min(),inplace=True)
test_set['sensor2_rolling60'].where(test_set['sensor2_rolling60'].isna()==False,test_set['sensor2_rolling60'].min(),inplace=True)
```

```python
def outlier_removal(df,columns):
    df_ = df.copy()
    for col in columns:
        high = df_[col].quantile(0.98)
        low = df_[col].quantile(0.02)
        df_.loc[df_[col]>high,col]=high
        df_.loc[df_[col]<low,col]=low
    return df_

test_set = outlier_removal(test_set,['sensor1','sensor2'])
```
